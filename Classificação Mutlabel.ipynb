```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.multioutput import ClassifierChain
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Musica.csv")
df.head(10)
```

```python
#separando as classes e previssores
y = df.iloc[:,0:6].values #selecionado todas as linhas e as 5 primeiras colunas
x = df.iloc[:,6:].values #para as colunas de previssores é da coluna 6 ate o fim do data frame
```

```python
#padronizando os previsores
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
print(x)
```

```python
#separando o DF
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=42)
```

```python
#maquina de vetor de suporte
svm = SVC()
classifier_chain = ClassifierChain(estimator=svm, random_state= 42)
classifier_chain.fit(x_train, y_train)
```

```python
#avaliando performance do modelo treinado
y_pred = classifier_chain.predict(x_test)
loss = hamming_loss(y_test, y_pred)
print(f'Hamming Loss: {loss}')

#o modelo apresentou um erro de apenas 19% no 
```

```python
#testando outras metricas
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, jaccard_score
f1_micro = f1_score(y_test, y_pred, average= 'micro')
f1_macro = f1_score(y_test, y_pred, average= 'macro')
subset_acc = accuracy_score(y_test, y_pred)
jaccard = jaccard_score(y_test, y_pred, average='samples')
print(f' F1 micro: {f1_micro}')
print(f' F1 macro: {f1_macro}')
print(f' Accurancy Score: {subset_acc}')
print(f' Jaccard Score: {jaccard}')
precision_micro = precision_score(y_test, y_pred, average="micro")
recall_micro = recall_score(y_test, y_pred, average="micro")
precision_macro = precision_score(y_test, y_pred, average="macro")
recall_macro = recall_score(y_test, y_pred, average="macro")
print(f' precision_micro: {precision_micro}')
print(f' precision_macro: {precision_macro}')
print(f' Recall micro: {recall_micro}')
print(f' Recall macro: {recall_macro}')
```
